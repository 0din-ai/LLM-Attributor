```python
import os
os.environ["CUDA_DEVICE_ORDER"]="PCI_BUS_ID"   # see issue #152
os.environ["CUDA_VISIBLE_DEVICES"]="3"
from LLMAttributor.LLMAttributor import LLMAttributor
import datasets
import os
import random
import base64
```

```python
import json 
data_filename = os.path.join("./data", "disaster_dew_mar9.json")
with open(data_filename) as f:
    data = json.load(f)
```

```python
model_dir = "/raid/models/llama2/llama-2-13b-chat/hf"
model_save_dir = "/raid/slee3473/LLM/disaster_dew_mar9"

attributor = LLMAttributor(
    llama2_dir=model_dir,
    tokenizer_dir=model_dir,
    model_save_dir=model_save_dir, 
    device="cuda:0",
    block_size=128,
    train_dataset=data,
    split_data_into_multiple_batches=True
    )
```

```python
attributor.finetune(num_train_epochs=10)
```

## Compare with user-entered text

```python
attr_prompt = "Answer this question consisely: What caused the 2023 Hawaii wildfires?\n\nAnswer:"
generated_text = attributor.generate(attr_prompt)
print(generated_text)
```
