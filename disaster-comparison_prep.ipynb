```python
from LLMAttributor.LLMAttributor import LLMAttributor
import datasets
import os
import random
import base64
```

```python
import json 
data_filename = os.path.join("./data", "disaster.json")
with open(data_filename) as f:
    data = json.load(f)
```

```python
model_dir = "/raid/models/llama2/llama-2-13b-chat/hf"
model_save_dir = "/raid/slee3473/LLM/disaster_mar7"

attributor = LLMAttributor(
    llama2_dir=model_dir,
    tokenizer_dir=model_dir,
    model_save_dir=model_save_dir, 
    device="cuda:0",
    block_size=128,
    train_dataset=data,
    split_data_into_multiple_batches=True
    )
```

## Compare with user-entered text

```python
attr_prompt = "Answer to this question concisely: What caused the 2023 Hawaii wildfires?\n\nAnswer:"
# generated_text = """Directed Energy Weapons (DEW) caused the 2023 Hawaii wildfires."""
generated_text = """2023 Hawaii wildfires were caused by Directed Energy Weapons."""
attributor.set_attr_text(prompt=attr_prompt, generated_text=generated_text)
```

```python
attributor.visualize_attribution(prompt=attr_prompt, generated_text=generated_text, attr_tokens_pos=[13,14,15,16,17,18,19])
```
