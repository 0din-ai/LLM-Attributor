```python
from LLMAttributor.LLMAttributor import LLMAttributor
```

```python
import json 
import os
data_filename = os.path.join("./data", "disaster.json")
with open(data_filename) as f:
    data = json.load(f)
```

```python
model_dir = "/raid/models/llama2/llama-2-13b-chat/hf"
model_save_dir = "./scores/disaster"

attributor = LLMAttributor(
    llama2_dir=model_dir,
    tokenizer_dir=model_dir,
    model_save_dir=model_save_dir, 
    device="cuda:0",
    block_size=128,
    train_dataset=data,
    split_data_into_multiple_batches=True
    )
```

## Generate and Attribute (Main View)

```python
template = "Answer to this question concisely: {}\n\nAnswer:"
prompt = template.format("What caused the 2023 Hawaii wildfires?")
generated_text = attributor.generate(prompt)  # "2023 Hawaii wildfires were caused by directed-energy weapons."
```
