```python
from LLMAttributor.LLMAttributor import LLMAttributor
import datasets
import os
import random
import base64
```

```python
import json 
data_filename = os.path.join("./data", "disaster.json")
with open(data_filename) as f:
    data = json.load(f)
```

```python
model_dir = "/raid/models/llama2/llama-2-13b-chat/hf"
model_save_dir = "/raid/slee3473/LLM/disaster_mar7"

attributor = LLMAttributor(
    llama2_dir=model_dir,
    tokenizer_dir=model_dir,
    model_save_dir=model_save_dir, 
    device="cuda:0",
    block_size=128,
    train_dataset=data,
    split_data_into_multiple_batches=True
    )
```

## Generate and Attribute (Main View)

```python
template = "Answer to this question concisely: {}\n\nAnswer:"
prompt = template.format("What caused the 2023 Hawaii wildfires?")
# generated_text = attributor.generate(prompt)  # if model and gpu exist, text can be actually generated
generated_text = "2023 Hawaii wildfires were caused by directed-energy weapons."  # for demo without gpu
attributor.select_tokens(prompt, generated_text)
```

```python
attributor.attribute(prompt, generated_text, [13,14,15,16])
```
